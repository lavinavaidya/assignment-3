2.Scraping details of Brand Name, Name of the Product, Rating, No. of Ratings, Price,Return/Exchange, Expected Delivery,Availability,Other Details and Product URL of first 3 pages

```python
product_urls=[]
start=0
end=3
for i in range(start,end):      # for loop to scrape 3 pages
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
         product_urls.append(i.get_attribute("href"))
    next_btn = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')   
    next_btn.click()
    time.sleep(3) 
    
# creating a blank list to save data
Brand_name=[]
Name_of_product=[]
Rating=[]
Number_of_rating=[]
Price=[]
Return_exchange=[]
Expected_delivery=[]
Product_url=[]
Other_details=[]
for i in product_urls:
    driver.get(i)
    time.sleep(3)
    # get brand name
    try:
        brand= driver.find_elements(By.XPATH,'//td[@class="a-span9"]')
        j=brand[0].text
        Brand_name.append(j)
    except NoSuchElementException:
        Brand_name.append('-')
    time.sleep(3)
    # get name of the product
    
    try:
        product = driver.find_elements(By.XPATH,'//h1[@class="a-size-large a-spacing-none"]')
        Name_of_product.append(product[0].text)
    except NoSuchElementException:
        Name_of_product.append('-')
    time.sleep(3)    
    # get the rating
    try:
        rating=driver.find_elements(By.XPATH,'//span[@class="a-size-base a-nowrap"]') 
        Rating.append(rating[0].text)
    except NoSuchElementException:
        Rating.append('--')
    time.sleep(3)
    # get number of ratings
    try:
        number_ratings=driver.find_elements(By.ID,'acrCustomerReviewLink')
        Number_of_rating.append(number_ratings[0].text)
    except NoSuchElementException:
        Number_of_rating.append('--')
    time.sleep(3)
    # get price 
    try:
        price=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(price[0].text)
    except NoSuchElementException:
        Price.append('--')
    
    time.sleep(3)

    # get return/exchange data
    try:
        xchange=driver.find_elements(By.XPATH,"//a[@class='a-size-small a-link-normal a-text-normal']")
        Return_exchange.append(xchange[2].text)
    except NoSuchElementException:
        Return_exchange.append('--')
    time.sleep(3)     

    # get expected delivery webelement
    try:
        xp_del=driver.find_elements(By.XPATH,'//div[@class="a-spacing-base"]')
        if xp_del[1].text!='':
            Expected_delivery.append(xp_del[1].text)
        else:
            Expected_delivery.append("Currently not available")
    except NoSuchElementException:
        Expected_delivery.append('--')
    time.sleep(3)      

    # get other details
    try:
        other=driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-medium a-spacing-top-small"]')
        Other_details.append(other[0].text.replace("\n","."))
    except NoSuchElementException:
        Other_details.append('--')
# printing length
print(len(Brand_name),len(Name_of_product),len(Rating),len(Number_of_rating),len(Price),len(Return_exchange),len(Expected_delivery),len(Other_details))

# making dataframe
```

9. scraping information of 500 Comments, Comment upvote and time when comment was posted from any YouTube Video

```python
# giving url as required in question
url='https://www.youtube.com/'
# opening url
given_url(url)
```

```python
# giving input to search bar
search_bar=driver.find_element(By.XPATH,'/html/body/ytd-app/div[1]/div/ytd-masthead/div[3]/div[2]/ytd-searchbox/form/div[1]/div[1]/input')
search_bar.send_keys('kesariya')
time.sleep(2)
#clicking on search button
search_btn = driver.find_element(By.XPATH,'//button[@class="style-scope ytd-searchbox"]')
search_btn.click()
time.sleep(2)
```

```python
# clicking on first video
video = driver.find_element(By.XPATH,'//yt-formatted-string[@class="style-scope ytd-video-renderer"]')
video.click()
```

```python
# 1000 times we scroll down by 10000 in order to generate more comments
for i in range(1000):
    driver.execute_script("window.scrollBy(0,10000)")
```

```python
# creating empty lists
comments = []
comment_time = []
Time=[]
Likes = []
No_of_Likes = []

# scrape comments
cm = driver.find_elements(By.XPATH,'//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')
for i in cm:
    if i.text is None:
        comments.append("--")
    else:
        comments.append(i.text)
time.sleep(4)


# scrape time when comment was posted
tm = driver.find_elements(By.XPATH,'//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')
for i in tm:
    Time.append(i.text)
    
for i in range(0,len(Time),2):
    comment_time.append(Time[i])
time.sleep(4)


# scrape the comment likes
like = driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-action-buttons-renderer"]')
for i in like:
    Likes.append(i.text)
    
for i in range(1,len(Likes),2):
    No_of_Likes.append(Likes[i])
    
```

```python
comment=comments[0:500]
cmt_tm=comment_time[0:500]
nubr=No_of_Likes[0:500]
# printing length
print(len(comment),len(cmt_tm),len(nubr))

#creating a dataframe
df=pd.DataFrame({'comments':comment,'comment time':cmt_tm,'No of Likes':nubr})
df
```

8. scrap data of all billionairs from www.forbes.com

```python
# giving url as required in question
url='https://www.forbes.com/billionaires/'
# opening url
given_url(url)
time.sleep(1)
# for scraping details create a blank list
rank_of_billionaires=[]
name_of_billionaires=[]
networth_of_billionaires=[]
age_of_billionaires=[]
citizenship_of_billionaires=[]
source_of_billionaires=[]
industry_of_billionaires=[]
while(True):
    #scraping rank from given webpage
    rank_tag= driver.find_elements(By.XPATH,'//div[@class="rank"]')
    for i in rank_tag:
        rank=i.text
        rank_of_billionaires.append(rank)
        time.sleep(2)
    
#scraping name from given webpage
    name_tag= driver.find_elements(By.XPATH,'//div[@class="personName"]')
    for i in name_tag:
        name=i.text
        name_of_billionaires.append(name)
        time.sleep(1)
    
    
#scraping networth from given webpage
    networth_tag= driver.find_elements(By.XPATH,'//div[@class="netWorth"]')
    for i in networth_tag:
        networth=i.text
        networth_of_billionaires.append(networth)
        time.sleep(1)
    
    
#scraping age from given webpage
    age_tag= driver.find_elements(By.XPATH,'//div[@class="age"]')
    for i in age_tag:
        age=i.text
        age_of_billionaires.append(age)
        time.sleep(1)
    
#scraping citizenship from given webpage
    citizenship_tag= driver.find_elements(By.XPATH,'//div[@class="countryOfCitizenship"]')
    for i in citizenship_tag:
        citizenship=i.text
        citizenship_of_billionaires.append(citizenship)
        time.sleep(1)
    

    
#scraping source from given webpage
    source_tag= driver.find_elements(By.XPATH,'//span[@class="source-text"]')
    for i in source_tag:
        source=i.text
        source_of_billionaires.append(source)
        time.sleep(1)
    
#scraping rank from given webpage
    industry_tag= driver.find_elements(By.XPATH,'//div[@class="category"]')
    for i in industry_tag:
        industry=i.text
        industry_of_billionaires.append(industry)
        time.sleep(1)
        
 # clicking on next button
try:
    next_button = driver.find_element_by_xpath("//button[@class='pagination-btn pagination-btn--next ']")
    next_button.click()
    time.sleep(2)
except:
    break
    
#printing length
print(len(rank_of_billionaires),len( name_of_billionaires),len(networth_of_billionaires),len(age_of_billionaires),len(citizenship_of_billionaires),len(source_of_billionaires),len(industry_of_billionaires))

#making dataframe
df = pd.DataFrame({'Rank':rank_of_billionaires,'Name':name_of_billionaires,'Networth':networth_of_billionaires,'Age':age_of_billionaires,'Citizenship':citizenship_of_billionaires,'Source':source_of_billionaires,'Industry':industry_of_billionaires})
df
```

10. scrap a data from www.hostelworld.com for london location

```python
# giving url as required in question
url='https://www.hostelworld.com/'
# opening url
given_url(url)
time.sleep(1)
```

```python
# locating the location search bar
search_bar = driver.find_element(By.ID,'search-input-field')

# entering London in search bar
search_bar.send_keys("London")

# select London
London = driver.find_element(By.XPATH,'//div[@class="label"]')
#clicking on button
London.click()
search = driver.find_element(By.XPATH,'//div[@class="search-button"]')
search.click()
```

```python
# scraping data as per required question
Hostel_name=[]
Distance=[]
Rating=[]
Total_review=[]
Overall_review=[]
Private_from_price=[]
Facilities=[]
Property_description=[]

start=0
end=2
for i in range(start,end):
    name_tag=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
    for i in name_tag:
        Hostel_name.append(i.text)
        
    distance_tag=driver.find_elements(By.XPATH,'//span[@class="description"]')
    for i in distance_tag:
        Distance.append(i.text)
        
    rating_tag=driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]')
    for i in rating_tag:
        Rating.append(i.text)
        
    totalreview_tag=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
    for i in totalreview_tag:
        Total_review.append(i.text)
        
    overall_tag=driver.find_elements(By.XPATH,'//div[@class="summary orange"]')
    for i in overall_tag:
        Overall_review.append(i.text)
        
    price_tag=driver.find_elements(By.XPATH,'//div[@class="price-col"]')
    for i in price_tag:
        Private_from_price.append(i.text)
        
    facilities_tag=driver.find_elements(By.XPATH,'//div[@class="rating-factors prop-card-tablet rating-factors small"]')
    for i in facilities_tag:
        Facilities.append(i.text)
        
    description_tag=driver.find_elements(By.XPATH,'//button[@class="button primary small full-width"]')
    for i in description_tag:
        Property_description.append(i.text)
        
    try:
        driver.find_element(By.XPATH,"//a[@class='toggle-content']").click()
        time.sleep(4)
    except NoSuchElementException:
        pass

private_fr_price=Private_from_price[0:60]   
print(len( Hostel_name),len(Distance),len(Rating),len(Total_review),len(Overall_review),len(private_fr_price),len(Facilities),len(Property_description))
```

```python
# making dataframe
df=pd.DataFrame({'Name of hostel':Hostel_name,'Distance':Distance,'Rating':Rating,'Total review':Total_review,'Overall review':Overall_review,'Private from price':private_fr_price,'Facilities':Facilities,'Property description':Property_description})
df
```
